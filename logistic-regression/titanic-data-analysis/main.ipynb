```python
sex = pd.get_dummies(data['Sex'],drop_first=True)
sex.head(5)
```

```python
embark = pd.get_dummies(data['Embarked'],drop_first=True)
embark.head()
```

```python
pcl = pd.get_dummies(data['Pclass'],drop_first=True)
pcl.head()
```

```python
data = pd.concat([data,sex,embark,pcl],axis=1)
data.head()
```

```python
data.drop("Cabin", axis=1)
```

```python
data.drop(['Pclass','Sex','Embarked','Cabin','PassengerId','Name','Ticket'],axis=1, inplace=True)
```

```python
data.head()
```

### Train Data

```python
X = data.drop('Survived',axis=1)
y = data['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
logmodel = LogisticRegression(solver='liblinear')
```

```python
logmodel.fit(X_train, y_train)
```

```python
predections = logmodel.predict(X_test)
```

```python
classification_report(y_test, predections)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, predections)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, predections)
```
